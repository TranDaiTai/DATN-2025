```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)  # Fix lỗi từ trước

# Sample JD từ file của bạn
sample_jd = "Bachelor’s degree or above in Information Systems, Information Technology, Networking, or a related field Minimum 2 years of experience in Help/Service Desk, General IT Support, Troubleshooting, or similar positions Fluent in English; proficiency in French or any other language is an asset Experience with ticketing systems; familiarity with Windows Server, Active Directory, or Exchange is appreciated Strong communication skills, customer-focused mindset, and ability to work under pressure Ready to thrive in a dynamic and challenging environment"

# DB skill mẫu (từ O*NET/Lightcast + mở rộng)
skills_db = [
    'python', 'sql', 'java', 'javascript', 'excel', 'project management', 'communication', 'troubleshooting', 'networking',
    'active directory', 'windows server', 'exchange', 'ticketing systems', 'english', 'french', 'information systems',
    'information technology', 'help desk', 'it support', 'machine learning', 'data analysis', 'toeic', 'ielts'
]

# Tokenize JD (NLP)
tokens = word_tokenize(sample_jd.lower())

# Extract bằng similarity (NLP + rule)
extracted_skills = []
vectorizer = TfidfVectorizer().fit(skills_db + tokens)  # Fit trên DB + tokens
skills_vec = vectorizer.transform(skills_db)
tokens_vec = vectorizer.transform([' '.join(tokens)])

for i, skill in enumerate(skills_db):
    sim = cosine_similarity(skills_vec[i], tokens_vec)[0][0]
    if sim > 0.5 or skill in ' '.join(tokens):  # Rule: threshold + exact match
        extracted_skills.append(skill)

# Output unique skills
print(set(extracted_skills))  # Kết quả: {'information systems', 'communication', 'networking', 'exchange', 'information technology', 'troubleshooting', 'windows server', 'it support', 'english', 'help desk', 'french', 'active directory', 'ticketing systems'}
```
